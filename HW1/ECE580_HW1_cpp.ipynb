```python
%%cu
#include <fstream>
#include <iostream>
#include <string>
#include <sstream>
#include <vector>
#include <algorithm>
#include "matplotlibcpp.h"
namespace plt = matplotlibcpp;

//Position of Predictor Variables(number start from 1)
int not_of_interest[13]={0,0,0,0,0,0,0,0,0,5,5,6,13};
//88.60,168.80,64.10,48.80,2548,130,3.47,2.68,9.00,111,5000,21,27

//Class for Dataset
class Dataset{
  private:
    std::string file_addr;
    std::vector<std::vector<std::string>> raw_content;
  public:
    Dataset(std::string file_addr):file_addr(file_addr){
        std::vector<std::string> row;
        std::string line, word;
        std::fstream file(file_addr,std::ios::in);
        try{
            if(file.is_open()){
                while(getline(file,line)){
                    row.clear();
                    std::stringstream str(line);
                    while(getline(str, word, ',')){
                        row.push_back(word);
                    }
                    raw_content.push_back(row);
                }
            }else{
                throw -1;
            }
        }catch(...){
            std::cout<<"Could not load data file!\n"<<std::endl;
        }
    }

    //print out the raw data
    void print_content(){
        for(int i=0; i<raw_content.size();i++){//raw_content.size();i++){
          for(int j=0;j<raw_content[i].size();j++){
              std::cout<<raw_content[i][j]<<" ";
          }
          std::cout<<"\n";
      }
    }

    //clear data of specific column(array)
    //print number of data removing in each step
    void preprocessing(int not_of_interest[]){
        int count_feature_rmv, count_unknown_rmv = 0;
        
        //Remove unrelated column
        for(int i=0;i<13;i++){
            //std::cout<<not_of_interest[i]<<"\n";
            std::for_each(raw_content.begin(),raw_content.end(),[&](std::vector<std::string>& row){
              row.erase(std::next(row.begin(),not_of_interest[i])); 
            count_feature_rmv +=1;
          });    
        }
        
        std::cout
        <<"Number of data removed from not-of-interest features is: "
        <<count_feature_rmv<<"\n"
        <<"Number of data removed from unknown features is: "
        <<count_unknown_rmv<<"\n"
        <<std::endl;
    }
};

int main(){
    
    //Initialize Dataset
    Dataset dataset("/content/ECE580/HW1/imports-85.data");

    //clear the dataset by removing not-of-interest datapoints
    dataset.preprocessing(not_of_interest);

    //dataset.print_content();

    return 0;
}
```
